```python
# Imports needed to complete this assignment
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#Read the CSV file into a dataframe
ibm = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
ibm.head()
```

```python
#1. SHOW A BREAKDOWN OF DISTANCE FROM HOME BY JOB ROLE AND ATTRITION.

dist_by_job_and_attrition = ibm.groupby(['DistanceFromHome','Attrition', 'JobRole']).agg({'JobRole': ['count']})
dist_by_job_and_attrition.head(25)
```

```python
#VISUALIZATION

df_active=ibm.groupby(["JobRole", "Attrition"])["DistanceFromHome"].nunique().reset_index()
plt.figure(figsize=(15,10))
sns.barplot(x="DistanceFromHome", y="JobRole", hue="Attrition", data=df_active)
plt.title("BREAKDOWN OF DISTANCE FROM HOME BY JOB ROLE AND ATTRITION")
plt.xlabel("Distance From Home")
plt.ylabel("Job Role")
```

```python
#2. COMPARE THE AVERAGE MONTHLY INCOME BY EDUCATION AND ATTRITION.

avg_month_income_by_edu_and_attrition = ibm.groupby(['Education','Attrition']).agg({'MonthlyIncome': ['mean']})
avg_month_income_by_edu_and_attrition
```

```python
#VISUALIZATION

avg_month_income_by_edu_and_att = ibm.groupby(["Education", "Attrition"])["MonthlyIncome"].mean().reset_index()
plt.figure(figsize=(15,10))
sns.barplot(x="Education", y="MonthlyIncome", hue="Attrition", data = avg_month_income_by_edu_and_att)
plt.title("AVERAGE MONTHLY INCOME BY EDUCATION AND ATTRITION")
plt.xlabel("Education")
plt.ylabel("MonthlyIncome")
```
